```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import random
import os

import yaml
import cv2
import shutil
import tqdm
import glob
import torch
from ultralytics import YOLO
```

```python
path = os.getcwd()
img_path = os.path.join(path,"img_data")
images_path = os.path.join(img_path,"All_img")
annotations_path  = os.path.join(img_path,"All_img_Label")
```

```python
yaml_path = os.path.join(path,'data.yaml')
```

```python
with open(yaml_path,'r') as f:
    yaml_file = yaml.load(f, Loader=yaml.FullLoader)
    display(yaml_file)
```

```python
# load pre-trained model
model = YOLO("yolov8m.pt")
model.train(data=yaml_path, epochs = 200, patience = 30, imgsz = 256)
```

```python
# model = YOLO("C:/Users/USER/runs/detect/train27/weights/best.pt")
List = {"Top":0, "Outer":1, "Pants":2, "Onepiece":3, "Skirt":4}
Count = {"Top":0, "Outer":0, "Pants":0, "Onepiece":0, "Skirt":0}
for now in List:
    test_loc = os.path.join(os.path.join(os.path.join(img_path,"img_data"),"category_classification"),now)
    results = model.predict(source = test_loc,save=True)
    for result in results:
        if List[now] in result.boxes.cls:
            Count[now]+=1
for now in List:
    print(now,": ",Count[now]/200*100,"%")
```
